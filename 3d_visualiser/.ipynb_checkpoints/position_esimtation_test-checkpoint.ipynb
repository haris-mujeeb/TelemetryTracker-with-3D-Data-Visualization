# Load Sensor Data

```python
from sensor_data_loadder import SensorData
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

DATA_DIR = './walks/outside_elab/new/'
DATA_FILE = 'trax_2025_4_25_3_37_16_#28.csv'

sensorData = SensorData(data_dir=DATA_DIR, data_file=DATA_FILE)
```

# Orientation Estimation using Accelerometer and Magnetometer Raw Data

```python
from orientation_estimation import KalmanOrientationEstimator
import numpy as np
import matplotlib.pyplot as plt

initial_covariance = np.eye(3) * 0.1**2  # Example values for position/angle uncertainty
process_noise = np.diag([0.01**2, 0.01**2, 0.02**2])  # Accelerometer, Gyroscope, Magnetometer
measurement_noise = np.diag([0.2**2, 0.2**2, 0.3**2])  # Accelerometer, Gyroscope, Magnetometer
alpha = 0.7

kalmanFilter = KalmanOrientationEstimator(
    sensorData.acc_raw_data,
    sensorData.mag_raw_data,
    sensorData.gyro_raw_data,
    sensorData.timestamps_in_ms,
    initial_covariance=initial_covariance,
    process_noise=process_noise,
    measurement_noise=measurement_noise,
    alpha=alpha
)

roll_kf, pitch_kf, yaw_kf = kalmanFilter.compute_angles()
```

# Positon Estimtaion
## Estimating Velocities from IMU

```python
from pymap3d import geodetic2enu

# Example: GPS raw data: (N, 3) where each row is [lat, lon, alt]
# gps_raw_data = np.array([[lat1, lon1, alt1], [lat2, lon2, alt2], ..., [latN, lonN, altN]])

lat0, lon0, alt0, NedNorthVel0, NedEastVel0  = sensorData.gps_raw_data[0]  # Origin for ENU conversion

# Initialize arrays to store ENU coordinates
e = np.zeros(sensorData.gps_raw_data.shape[0])  # East
n = np.zeros(sensorData.gps_raw_data.shape[0])  # North
u = np.zeros(sensorData.gps_raw_data.shape[0])  # Up

# Convert each GPS point (lat, lon, alt) to ENU
for i, (lat, lon, alt, nedNorthVel, nedEastVel) in enumerate(sensorData.gps_raw_data):
    e[i], n[i], u[i] = geodetic2enu(lat, lon, alt, lat0, lon0, alt0)

# Stack the results into a single (N, 3) array for ENU positions
gps_enu = np.stack([e, n, u], axis=1)

# For NED (North-East-Down) convention:
# X: North, Y: East, Z: Down
# For ENU (East-North-Up) convention:
# X: East, Y: North, Z: Up
```

```python
from scipy.spatial.transform import Rotation as Rscipy

# Combine roll, pitch, yaw into Euler angle array
euler_angles = np.vstack((roll_kf, pitch_kf, yaw_kf)).T  # Shape: (N, 3)

# Convert Euler angles to rotation matrices
rotations = Rscipy.from_euler('xyz', euler_angles)
rot_matrices = rotations.as_matrix()

# --- Prepare arrays ---
acc_world_ned = np.zeros_like(sensorData.acc_raw_data)
acc_world_enu = np.zeros_like(sensorData.acc_raw_data)

# --- Rotate acceleration to NED and convert to ENU ---
for i in range(len(sensorData.acc_raw_data)):
    acc_world_ned[i] = rot_matrices[i] @ sensorData.acc_raw_data[i]  # body → NED
    acc_world_enu[i, 0] = acc_world_ned[i, 1]              # East = NED Y
    acc_world_enu[i, 1] = acc_world_ned[i, 0]              # North = NED X
    acc_world_enu[i, 2] = -acc_world_ned[i, 2]             # Up = -Down

# --- Integration helper ---
def integrate(acc_data, time_deltas):
    vel = np.zeros_like(acc_data)
    pos = np.zeros_like(acc_data)
    for i in range(1, len(acc_data)):
        vel[i] = vel[i-1] + acc_data[i-1] * time_deltas[i-1]
        pos[i] = pos[i-1] + vel[i-1] * time_deltas[i-1]
    return vel, pos

# --- Integrate all three cases ---
vel_raw, pos_raw = integrate(sensorData.acc_raw_data, sensorData.time_deltas)
vel_ned, pos_ned = integrate(acc_world_ned, sensorData.time_deltas)
vel_enu, pos_enu = integrate(acc_world_enu, sensorData.time_deltas)

# --- Plotting ---
fig, axs = plt.subplots(2, 2, figsize=(15, 10), sharex='col')

# Velocity plots (East / x-axis and North / y-axis)
axs[0, 0].plot(vel_raw[:, 0], label='Raw', linestyle='--', alpha=0.6)
axs[0, 0].plot(vel_ned[:, 0], label='NED Rotated', linestyle='-.')
axs[0, 0].plot(vel_enu[:, 0], label='ENU Converted', linestyle='-')
axs[0, 0].set_title('Velocity - East')
axs[0, 0].legend()
axs[0, 0].grid(True)

axs[1, 0].plot(vel_raw[:, 1], label='Raw', linestyle='--', alpha=0.6)
axs[1, 0].plot(vel_ned[:, 1], label='NED Rotated', linestyle='-.')
axs[1, 0].plot(vel_enu[:, 1], label='ENU Converted', linestyle='-')
axs[1, 0].set_title('Velocity - North')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Position plots (East / x-axis and North / y-axis)
axs[0, 1].plot(pos_raw[:, 0], label='Raw', linestyle='--', alpha=0.6)
axs[0, 1].plot(pos_ned[:, 0], label='NED Rotated', linestyle='-.')
axs[0, 1].plot(pos_enu[:, 0], label='ENU Converted', linestyle='-')
axs[0, 1].set_title('Position - East')
axs[0, 1].legend()
axs[0, 1].grid(True)

axs[1, 1].plot(pos_raw[:, 1], label='Raw', linestyle='--', alpha=0.6)
axs[1, 1].plot(pos_ned[:, 1], label='NED Rotated', linestyle='-.')
axs[1, 1].plot(pos_enu[:, 1], label='ENU Converted', linestyle='-')
axs[1, 1].set_title('Position - North')
axs[1, 1].legend()
axs[1, 1].grid(True)

plt.suptitle('Comparison of Raw, NED-Rotated, and ENU-Converted Acceleration Integrations')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

### Result
It can be observed from the figure that the estimated position is only valid for very small interval of time due to high integration errors.
